## 设置坐标轴

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-3,3,50)
y1=2*x+1
y2=x**2

plt.figure()      #建立窗口
plt.plot(x,y2)
plt.plot(x,y1,color='red',linewidth=2.0,linestyle='--')

plt.xlim(-1,2)
plt.ylim(-2,3)
plt.xlabel('I am X')
plt.ylabel('I am Y')

new_ticks=np.linspace(-1,2,5)
print(new_ticks)
plt.xticks(new_ticks)

plt.yticks([-2,-1.3,0.2,1.22],[r'$really\ \alpha\ good$',r'$normal$',r'$roght$',r'$good$'])  #  &为转为数学符号所用，\ 用于转义空格

#gca=get current axis
ax=plt.gca()
ax.spines['right'].set_color('none')     #去掉边框
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')    #重新设置坐标轴
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data',-1))    #挪动坐标轴位置
ax.spines['left'].set_position(('data',0))

plt.show()
```
