## figure的使用

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-1,3,100)
y1=2*x+1
y2=x**2

plt.figure()
plt.plot(x,y1)

plt.figure(figsize=(8,4))     #设置figure的尺寸size
plt.plot(x,y2)
plt.plot(x,y1,color='red',linewidth=1.0,linestyle='--')   # 设置颜色、线宽、线型

plt.show()
```
